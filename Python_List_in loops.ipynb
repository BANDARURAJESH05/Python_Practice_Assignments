### Question 1: Write a Python program to count the number of elements in a list without using len.


- `sample input : `

- `sample output :`
    - `result : 4`

```python
# Please write The solution Here for above question 1:
var = [10,11,12,13]
res = 0
for i in var:
  res = res +1
print(f"result: {res}")
```

### Question 2: Write a Python program to find the maximum and minimum values in a given list.

- `sample input : `
    - `var = [10,11,12,13,14,15]`
- `sample output :`
    - `result : max = 15 , min = 10`

```python
# Please write The solution Here for above question 2:
var = [10,11,12,13,14,15]
max_val = var[0]
min_val = var[0]
for i in var:
  if i>max_val:
    max_val = i
  if i< min_val:
    min_val=i
print(f"result :max = {max_val},min = {min_val}")



```

### Question 3: Write a Python program to remove the first occurrence of a specified element from a list.

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : [1,3,4,2,5]`
    

```python
# Please write The solution Here for above question 3:
var = [1,2,3,4,2,5]
num = int(input())
for i in var:
  if var[i] == num:
    del var[i]
    break
print(var)

```

### Question 4: Write a Python program to find the average of all the elements in a list.


- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 2.833 `

```python
# Please write The solution Here for above question 4:
var = [1,2,3,4,2,5]
length = len(var)
sum = 0
for i in var:
  sum=sum+i
average= sum/length
print(f"result: {round(average,3)}")
```

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
# Please write The solution Here for above question 7:
var = [134,121,155]
for i in var:
  t = i
  rev = 0
  while t>0:
    res= t%10
    rev=rev*10+res
    t = t//10
  if rev ==i:
    print(i,"is palindrome")
  else:
    print(i,"not a palindrome")
```

### Question 10:  Write a Python program to sort a list of strings in Ascii Order [Ascending order]


- `sample input : `

- `sample output :`
    - `result : ['a' , 'b' , 'c']`


```python
# Please write The solution Here for above question 10:

var = ['a' , 'c' , 'b']
var.sort()
print(var)
```

```python
var = ['a' , 'c' , 'b']
n=len(var)
for i in range(n):
  for j in range(0,n-i-1):
    if var[j]>var[j+1]:
      var[j],var[j+1]=var[j+1],var[j]
print(var)
```
